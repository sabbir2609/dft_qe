```python
import crystal_toolkit
import os
from pymatgen.core import Structure
```

```python
znfo_expt = Structure.from_file(
    os.path.join("structures", "Zn(FeO2)2.cif")
)
print(znfo_expt)
znfo_expt
```

```python
import matcalc
from matcalc.utils import UNIVERSAL_CALCULATORS
```

```python
calculator_pbe = matcalc.load_fp("pbe")
```

```python
relax_calc = matcalc.RelaxCalc(
    calculator_pbe,
    optimizer='FIRE',
    relax_atoms=True,
    relax_cell=True
)
```

```python
data = relax_calc.calc(znfo_expt)
```
